```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
```

```python
# 1
imagem_dinossauro = cv2.imread("C:\\Users\\ronal\\OneDrive\\Documentos\\Projetos\\visao-computacional\\tarefa01\\dinossauro.jpg")
```

```python
# 2
print(imagem_dinossauro)
```

```python
# 3
plt.imshow(imagem_dinossauro)
```

```python
# 4
imagem_dinossauro_cinza = cv2.cvtColor(imagem_dinossauro, cv2.COLOR_BGR2GRAY)
plt.imshow(imagem_dinossauro_cinza, cmap='gray')
```

```python
# 5
imagem_dinossauro_negativa = (255-imagem_dinossauro)
plt.imshow(imagem_dinossauro_negativa) 
```

```python
# 6
r, g, b = cv2.split(imagem_dinossauro)

plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(r, cmap='gray');plt.title("Vermelho")
plt.subplot(142);plt.imshow(g,cmap='gray');plt.title("Verde")
plt.subplot(143);plt.imshow(b,cmap='gray');plt.title("Azul")

imagem_dinossauro_mesclada = cv2.merge((r,g,b))
plt.subplot(144);plt.imshow(imagem_dinossauro_mesclada[:,:,::-1]);plt.title("Sobreposição dos canais RGB");
```

```python
# 7
imagem_dinossauro_HSV = cv2.cvtColor(imagem_dinossauro, cv2.COLOR_BGR2HSV)

h, s, v = cv2.split(imagem_dinossauro_HSV)

plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(h,cmap='gray');plt.title("H");
plt.subplot(142);plt.imshow(s,cmap='gray');plt.title("S");
plt.subplot(143);plt.imshow(v,cmap='gray');plt.title("V");
plt.subplot(144);plt.imshow(imagem_dinossauro_HSV);plt.title("Original");
```

```python
# 2
cv2.imwrite("dinossauro-cinza.jpg", imagem_dinossauro_cinza)
cv2.imwrite("dinossauro-negativa.jpg", imagem_dinossauro_negativa)
cv2.imwrite("dinossauro-HSV.jpg", imagem_dinossauro_HSV)
cv2.imwrite("dinossauro-mesclada.jpg", imagem_dinossauro_mesclada)
```
