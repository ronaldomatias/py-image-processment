```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
```

```python
# 1
imagem_dinossauro = cv2.imread("C:\\Users\\ronal\\OneDrive\\Documentos\\Projetos\\visao-computacional\\tarefa01\\dinossauro.jpg")
```

```python
# 2
r, g, b = cv2.split(imagem_dinossauro)

plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(r,cmap='gray');plt.title("Vermelho");
plt.subplot(142);plt.imshow(g,cmap='gray');plt.title("Verde");
plt.subplot(143);plt.imshow(b,cmap='gray');plt.title("Azul");
```

```python
# 3
plt.imshow(imagem_dinossauro)

histograma = cv2.calcHist([imagem_dinossauro], [0], None, [256],[0, 256])
plt.figure()
plt.title("Histograma")
plt.xlabel("Intensidade")
plt.ylabel("Pixels")
plt.plot(histograma)
plt.xlim([0, 256])
plt.show()
```

```python
# 4
canais = cv2.split(imagem_dinossauro)
cores = ('b', 'g', 'r')

plt.figure()
plt.title('Histograma RGB')
plt.xlabel('Intensidade')
plt.ylabel('Pixels')
for (canal, cor) in zip(canais, cores):
    hist = cv2.calcHist([canal], [0], None, [256], [0,256])
    plt.plot(hist, cor)
    plt.xlim([0,256])
plt.show()
```

```python
# 5
imagem_dinossauro_RGB = cv2.cvtColor(imagem_dinossauro, cv2.COLOR_BGR2GRAY)
imagem_dinossauro_equalizada = cv2.equalizeHist(imagem_dinossauro_RGB)

plt.figure()
plt.title("Histograma Equalizado")
plt.xlabel("Intensidade")
plt.ylabel("Pixels")
plt.hist(imagem_dinossauro_equalizada.ravel(), 256, [0,256])
plt.xlim([0,256])
plt.show()

plt.figure()
plt.title("Histograma Original")
plt.xlabel('Intensidade')
plt.ylabel('Pixels')
plt.hist(imagem_dinossauro_RGB.ravel(), 256, [0,256])
plt.xlim([0,256])
plt.show()
cv2.waitKey(0)

plt.imshow(imagem_dinossauro_equalizada, cmap="gray")
```

```python
# 6 e 7 - Circulando o olho do bicho.

imagem_dinossauro_redimensionada = imagem_dinossauro[0:350,0:750];

cv2.circle(imagem_dinossauro_redimensionada, (500,80), 30, (0, 0, 255), thickness=1, lineType=cv2.LINE_AA);
plt.imshow(imagem_dinossauro_redimensionada[:,:,::-1])
```

```python
Parameters cv2.circle: 
img (CvArr) – Image where the circle is drawn
center (CvPoint) – Center of the circle
radius (int) – Radius of the circle
color (CvScalar) – Circle color
thickness (int) – Thickness of the circle outline if positive, otherwise this indicates that a filled circle is to be drawn
lineType (int) – Type of the circle boundary, see Line description
shift (int) – Number of fractional bits in the center coordinates and radius value
```
